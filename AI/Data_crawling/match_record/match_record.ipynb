```python
def format_date(year, month, day):
    # 연도, 월, 일을 문자열로 변환하고, 월과 일이 한 자리수일 경우 앞에 '0'을 추가
    formatted_date = f"{year}{month:02d}{day:02d}"
    return formatted_date
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 각 날짜의 데이터를 저장할 리스트
games_data = []

for year in range(2019, 2025):
    for month in range(3, 11):
        # 요청에 필요한 데이터 설정
        url = f'https://statiz.sporki.com/schedule/?year={year}&month={month}'
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'}

        response = requests.post(url, headers=headers)  # 또는 requests.get() 사용

        res = requests.get(url, headers=headers)
        res.raise_for_status()

        html = response.text

        # 응답에서 데이터 추출
        soup = BeautifulSoup(html, "html.parser")

        # td 태그를 모두 찾음8
        tds = soup.find_all('td')

        for td in tds:
            day = td.find('span', class_='day')
            if day is None:
                continue
            else:
                day = int(day.text)

            games = td.find_all('li')

            for game in games:
                teams = game.find_all('span', class_='team')
                scores = game.find_all('span', class_='score')
                if len(scores) == 0:    # 우천 취소된 경기일 경우 pass
                    continue

                game_data = [format_date(year, month, day), teams[0].text, scores[0].text, teams[1].text, scores[1].text]
                games_data.append(game_data)
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)

match_result_df = pd.DataFrame(games_data, columns=['Date', 'Away team', 'Away team score', 'Home team', 'Home team score'])

match_result_df = match_result_df.dropna(axis=0)

match_result_df
```
