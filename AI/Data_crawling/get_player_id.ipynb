```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

team_codes = [6002, 10001, 9001, 5002, 11001, 12001, 2002, 1001, 7002, 3001, 9002]  # 두산, 키움, SK(19~20), LG, NC, KT, KIA, 삼성, 한화, 롯데, SSG(21~)
all_player_dict = {}  # 전체 선수 ID와 이름을 저장할 딕셔너리

for team_code in team_codes:
    for year in range(2019, 2025):
        if year > 2020 and team_code == 9001:
            continue

        # 요청에 필요한 데이터 설정
        url = f'https://statiz.sporki.com/team/?m=seasonBacknumber&t_code={team_code}&year={year}'
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'}

        response = requests.get(url, headers=headers)
        response.raise_for_status()

        html = response.text

        # 응답에서 데이터 추출
        soup = BeautifulSoup(html, "html.parser")

        for link in soup.find_all('a'):
            if '/player/?m=playerinfo&amp;p_no=' in str(link):
                player_no = re.search(r'\d+', str(link)).group()
                player_no = int(player_no)
                player_name = link.text
                all_player_dict[player_no] = player_name

print(all_player_dict)
```
